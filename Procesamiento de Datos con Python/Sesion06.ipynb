## Ejemplo 3: Excepciones y try except

```python
# Lista de entradas como cadenas, algunas de las cuales no son números válidos
lista_cadenas = ["5", "12", "error", "7", "tres", "10"]

# Lista vacía para guardar los números convertidos
numeros = []

# Bucle para convertir cada cadena a entero
for cadena in lista_cadenas:
    try:
        # Intenta convertir la cadena a entero
        numero = int(cadena)
        numeros.append(numero)
    except ValueError:
        # Maneja la excepción si la conversión falla
        print(f"No se pudo convertir '{cadena}' a entero.")

# Imprime la lista resultante
print("Números convertidos:", numeros)
```

```python
lista_2 = [1, 2, 3, 4, 5]

try:
    print(lista_2[10])
except:
    print("Ese numero esta fuera de rango")
    print("Mejor leamos este número")
    print(lista_2[2])
```

```python
def dividir(a, b):
    try:
        resultado = a / b
    except ZeroDivisionError:
        print("No se puede dividir por cero.")
    else:
        print(f"El resultado de la división es {resultado}")
    finally:
        print("Operación de división finalizada.")

# Probando la función
dividir(10, 2)  # División normal
dividir(10, 0)  # División por cero
```

```python
#-------------------------------------#
# Reto 2: Automatizando con for loops #
#-------------------------------------#
def potencia_de_dos(exponente):
    resultado = 2
    if exponente == 0:
      return 1
    for i in range(0, exponente - 1):
      resultado = resultado **2

    return resultado
print(resultado)
```

## Ejemplo 4: Concat con Series

```python
import pandas as pd
serie_1 = pd.Series([1,2,3], index = ['a', 'b', 'c'])
serie_2 = pd.Series([4,5,6], index = ['d', 'f', 'g'])
```

```python
#axis = 0 se esta cocatenando a lo largo del eje de las filas,  por lo tanto serie_1 y serie_2 se "apilaran" sobre la otra, extendiendo el DataFrame a lo largo de su eje vertical.
pd.concat([serie_1,serie_2], axis = 0) #cocatenar de forma vertical
```

```python
# axis =  1 se cocatena a lo largo del eje de las columnas horizontal
pd.concat([serie_1,serie_2], axis = 1) #como Data Frame
#tienen un valor Nan ya que "completa" la tabla
```

##Reto 3: Evitando errores con try except


```python
import pandas as pd
import numpy as np
def lista_de_strings_a_lista_de_floats(lista_de_strings):

    lista_de_floats = []

    for string in lista_de_strings:
      try:
        float_ = float(string)
        lista_de_floats.append(float_)
      except:
        lista_de_floats.append(np.nan)
    return lista_de_floats

lista_strings = ['1.2', '3.4', '5.6', 'abc', '7.8']

# Llamar a la función y almacenar el resultado
lista_floats = lista_de_strings_a_lista_de_floats(lista_strings)

# Imprimir los resultados
print("Lista original de strings:", lista_strings)
print("Lista convertida a floats:", lista_floats)
```
