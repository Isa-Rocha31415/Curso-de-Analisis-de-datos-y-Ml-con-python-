````python
import pandas as pd
import numpy as np
````

````python
def years_since_bestseller(valor):
  as_datetime = pd.to_datetime(valor, unit='ms')
  today = pd.to_datetime('today')
  difference_in_days = (today - as_datetime).days
  in_years = difference_in_days / 365
  return in_years
````

````python
df['published_date.numberLong'].apply(years_since_bestseller)
````

````python
def weeks_on_list_percentage_of_maximum(value, max_weeks_on_list):

    percentage = value * 100 / max_weeks_on_list
    as_string = f'{percentage:.2f}%'

    return as_string
````

````python
df['weeks_on_list.numberInt'].apply(weeks_on_list_percentage_of_maximum, args=(df['weeks_on_list.numberInt'].max(),))
````

Explicación con mas detalles sobre los parámetros más comunes del método `sort_values` en Pandas:

1. **by:** Especifica el nombre o la lista de nombres de las columnas por las cuales quieres ordenar. Si ordenas por más de una columna, Pandas primero ordenará por la primera columna indicada, luego dentro de esos resultados, ordenará por la siguiente columna, y así sucesivamente.

2. **axis:** Determina si ordenarás por filas o por columnas. Por defecto es `0`, lo que significa que ordena las filas. Si se establece en `1`, ordenará las columnas.

3. **ascending:** Un booleano o una lista de booleanos que indica si el ordenamiento debe ser ascendente (`True`) o descendente (`False`). Si `by` especifica más de una columna y `ascending` es una lista, cada booleano en la lista corresponderá a cada columna en `by`.

4. **inplace:** Si se establece en `True`, el ordenamiento se aplicará en el DataFrame original, modificándolo. Si es `False` (por defecto), Pandas devolverá una nueva instancia del DataFrame ordenado, dejando el original sin cambios.

5. **kind:** Especifica el algoritmo de ordenamiento a usar. Las opciones incluyen `quicksort` (por defecto), `mergesort`, `heapsort`, etc. Cada algoritmo tiene sus propias características en términos de rendimiento y estabilidad (si mantiene el orden relativo de los elementos iguales).

6. **na_position:** Indica dónde deben colocarse los valores `NaN`. Las opciones son `'first'` para colocar `NaN` al principio o `'last'` (por defecto) para colocarlos al final.

7. **ignore_index:** Si se establece en `True`, el índice del DataFrame se ignorará durante el ordenamiento y se generará un nuevo índice de 0 a N-1. Esto puede ser útil si el índice original ya no tiene relevancia después de ordenar los datos.

Por ejemplo, si tienes un DataFrame `df` y quieres ordenarlo por la columna `'edad'` de manera descendente y decidir que los valores `NaN` aparezcan al principio, podrías escribir:

```python
df.sort_values(by='edad', ascending=False, na_position='first')
```

Estos parámetros te dan mucha flexibilidad para ordenar tus datos de la manera que necesites.

El parámetro `ascending` en `sort_values` determina el orden en el que se ordenan los valores de la columna especificada:

1. **`ascending=True`:**
   - Cuando configuras `ascending=True`, estás indicando que quieres ordenar los datos en orden ascendente.
   - Esto significa que los valores más pequeños o más antiguos aparecerán primero.
   - Por ejemplo, si estás ordenando fechas con `ascending=True`, las fechas más antiguas aparecerán al principio de tu DataFrame.

2. **`ascending=False`:**
   - Al configurar `ascending=False`, los datos se ordenarán en orden descendente.
   - Esto significa que los valores más grandes o más recientes aparecerán primero.
   - Por ejemplo, al ordenar fechas con `ascending=False`, las fechas más recientes estarán al principio del DataFrame.

En el contexto de tu ejemplo, `df.sort_values('published_date.numberLong', ascending=True)` ordenará el DataFrame `df` basado en la columna `'published_date.numberLong'` en orden ascendente. Esto colocará los libros con las fechas de publicación más antiguas al principio del DataFrame. Si cambiaras a `ascending=False`, los libros con las fechas de publicación más recientes estarían al principio.
