## Ejemplo 2: Modificando listas

````python
# .append()
# .pop()
````

1. `append()`: Agrega un elemento al final de la lista.
   ```python
   lista = [1, 2, 3]
   lista.append(4)  # lista ahora es [1, 2, 3, 4]
   ```

2. `clear()`: Elimina todos los elementos de la lista.
   ```python
   lista.clear()  # lista ahora está vacía []
   ```

3. `copy()`: Retorna una copia de la lista.
   ```python
   nueva_lista = lista.copy()  # nueva_lista es una copia de lista
   ```

4. `count()`: Retorna el número de elementos con el valor especificado.
   ```python
   conteo = lista.count(2)  # conteo es el número de veces que aparece 2 en lista
   ```

5. `extend()`: Agrega los elementos de una lista (o cualquier iterable), al final de la lista actual.
   ```python
   lista.extend([5, 6])  # lista ahora es [1, 2, 3, 4, 5, 6]
   ```

6. `index()`: Retorna el índice del primer elemento con el valor especificado.
   ```python
   indice = lista.index(3)  # indice es la posición donde 3 aparece por primera vez en lista
   ```

7. `insert()`: Agrega un elemento en la posición especificada.
   ```python
   lista.insert(1, 'a')  # lista ahora es [1, 'a', 2, 3, 4]
   ```

8. `pop()`: Elimina el elemento en la posición especificada y lo retorna.
   ```python
   elemento = lista.pop(2)  # elemento es 2 y lista ahora es [1, 'a', 3, 4]
   ```

9. `remove()`: Elimina el primer ítem con el valor especificado.
   ```python
   lista.remove('a')  # lista ahora es [1, 2, 3, 4]
   ```

10. `reverse()`: Invierte el orden de la lista.
    ```python
    lista.reverse()  # lista ahora es [4, 3, 2, 1]
    ```

11. `sort()`: Ordena la lista.
    ```python
    lista.sort()  # lista ahora está ordenada [1, 2, 3, 4]
    ```

````python
# Funciones con objetos **kwargs
def crear_usuario(**kwargs):
    usuario = {
        "nombre": kwargs.get("nombre", "Nombre no proporcionado"),
        "edad": kwargs.get("edad", "Edad no proporcionada"),
        "email": kwargs.get("email", "Email no proporcionado")
    }
    return usuario

# Creando un diccionario de datos del usuario
datos_usuario = {"nombre": "Carlos", "edad": 28, "email": "carlos@example.com"}

# Pasando el diccionario como **kwargs
perfil = crear_usuario(**datos_usuario)
print(perfil)
````

````python
# Funciones con objeto diccionario simple
def analizar_datos(datos):
    # Supongamos que hace algún análisis complejo aquí
    resumen = {
        "número de elementos": len(datos),
        "claves": list(datos.keys())
    }
    return resumen

# Creando un diccionario con algunos datos
datos_muestra = {"dato1": 10, "dato2": 20, "dato3": 30}

# Pasando el diccionario como un objeto normal
resultado = analizar_datos(datos_muestra)
print(resultado)
````

````python
# Funciones con argumentos como funciones.

# Función para verificar si un valor es un entero
def es_entero(valor):
    return isinstance(valor, int)

# Función para verificar si un valor es un string
def es_string(valor):
    return isinstance(valor, str)

# Función que toma otra función y valores como argumentos
def verificar_tipo(funcion_verificacion, valor):
    if funcion_verificacion(valor):
        print(f"El valor '{valor}' pasa la verificación de {funcion_verificacion.__name__}.")
    else:
        print(f"El valor '{valor}' no pasa la verificación de {funcion_verificacion.__name__}.")

# Ejemplos de uso
verificar_tipo(es_entero, 100)  # Verifica si 100 es un entero
verificar_tipo(es_string, "Hola Mundo")  # Verifica si "Hola Mundo" es un string
verificar_tipo(es_entero, "Prueba")  # Verifica si "Prueba" es un entero
````

````python
# Funciónes Recursiva en Python
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n - 1)
print(factorial(5))  # 120
# print(factorial(959))  # 120
````

````python
# Solucion Reto Funciones
# a) Función numero_es_par
def numero_es_par(numero):

    if numero % 2 == 0:
        return True
    else:
        return False
````

````python
def test_funcion(funcion_a_probar, num_de_errores, contador, parametros=[], resultado_esperado=None):
    resultado_test = funcion_a_probar(*parametros)
    print(f'Test {contador}: Resultado esperado es `{resultado_esperado}`, obtuvimos `{resultado_test}`')
    if resultado_test != resultado_esperado:
        num_de_errores += 1

    return num_de_errores
````

````python
print("== Tests numero_es_par==\n")

contador_de_tests = 0
errores = 0

contador_de_tests += 1
errores = test_funcion(numero_es_par, errores, contador_de_tests, parametros=[2], resultado_esperado=True)
contador_de_tests += 1
errores = test_funcion(numero_es_par, errores, contador_de_tests, parametros=[3], resultado_esperado=False)
contador_de_tests += 1
errores = test_funcion(numero_es_par, errores, contador_de_tests, parametros=[0], resultado_esperado=True)
contador_de_tests += 1
errores = test_funcion(numero_es_par, errores, contador_de_tests, parametros=[127], resultado_esperado=False)
contador_de_tests += 1
errores = test_funcion(numero_es_par, errores, contador_de_tests, parametros=[-88], resultado_esperado=True)
contador_de_tests += 1
errores = test_funcion(numero_es_par, errores, contador_de_tests, parametros=[-1349], resultado_esperado=False)

print(f'\nErrores encontrados: {errores}')
````

````python
def sacar_porcentaje(muestra, total):
    return muestra * 100 / total

def crear_string_con_porcentaje_de_muestra_al_total(muestra, total, porcentaje):
    return f'{muestra} es el {porcentaje}% del número {total}'
````

````python
# Ejemplos de muestra y total
muestras = [50, 30, 20, 10, 5]
total = 100

# Calculando y formateando los resultados
resultado_1 = crear_string_con_porcentaje_de_muestra_al_total(muestras[0], total, sacar_porcentaje(muestras[0], total))
print(resultado_1)

resultado_2 = crear_string_con_porcentaje_de_muestra_al_total(muestras[1], total, sacar_porcentaje(muestras[1], total))
print(resultado_2)

resultado_3 = crear_string_con_porcentaje_de_muestra_al_total(muestras[2], total, sacar_porcentaje(muestras[2], total))
print(resultado_3)

resultado_4 = crear_string_con_porcentaje_de_muestra_al_total(muestras[3], total, sacar_porcentaje(muestras[3], total))
print(resultado_4)

resultado_5 = crear_string_con_porcentaje_de_muestra_al_total(muestras[4], total, sacar_porcentaje(muestras[4], total))
print(resultado_5)

muestra_adicional = 25
resultado_6 = crear_string_con_porcentaje_de_muestra_al_total(muestra_adicional, total, sacar_porcentaje(muestra_adicional, total))
print(resultado_6)
````
